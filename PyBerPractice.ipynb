```python
import matplotlib.pyplot as plt
import statistics
```

```python
savings = [0,100,150,200,250,300,500,600,300, 750,600,1000]
months= ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov','Dec']
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.bar(x_axis, y_axis, color='green',label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Creat the plot
plt.barh(x_axis,y_axis, color='magenta', label='Boston')
#invert y axis from Dec to Jan to jan - Dec
plt.gca().invert_yaxis()
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.legend()
```

```python
# create bar chart with object_oriented method
fig,ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')
ax.invert_yaxis() # don't have .gca() method
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend()
```

```python
# Create a scatterplot with plt.plot(); 
#add a lowercase "o" as a parameter inside the parentheses. This switches the plot from a line chart to a scatter plot
plt.plot(x_axis, y_axis, 'o', color='red')
```

```python
# Create scatterplot with plt.scatter()
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()
plt.xlim(0,50)
plt.legend()
```

```python
# list comprehension new_list=[expression for item in list of condition]
#e.g. s=[i*3 for i in y_axis]
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# Create scatter plot with object-oriented interface
fit,ax=plt.subplots()
ax.scatter(y_axis, x_axis, color='skyblue', s=[i*5 for i in y_axis],alpha=0.8, edgecolors='black',linewidth=2, label="Boston")
ax.legend()
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.invert_yaxis()
ax.set_xlim(0,50)
```

```python
#Assign 12 colors to 12 months
colors=["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values=(0,0,0,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8,8))
plt.pie(y_axis, 
        labels=x_axis, 
        autopct="%.1f%%",
        explode=explode_values, 
        colors=colors)

plt.show()
```

```python
# Add error bars to A Line Chart
#step1: set the x-axis to a list of strings for each month
x_axis =['Jan', 'Feb', 'Mar','April', 'May', 'June', 'July', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']

#step2: set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Add the x-axis, y-axis, and yerr=stdev to the errorbar()function
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add error bar with the object-oriented method
fig,ax=plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize =3)
plt.show()
```

```python
# Add errorbar to a bar chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# create error bar on bar chart with object-oriented approach
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# adjust ticks for horizontal bar chart
import numpy as np
```

```python
plt.barh(x_axis, y_axis) # plt.barh(y,x),plt.ylabel('y axis name'),plt.xlabel('x axis name')
plt.xticks(np.arange(0,51, step=5.0))
plt.gca().invert_yaxis()
```

```python
# Add minor ticks between the major ticks on the x- or y-axis
#import MultipleLocator
from matplotlib.ticker import MultipleLocator
```

```python
# Increase the size of the plot figure
fig,ax=plt.subplots(figsize=(8,8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))
ax.invert_yaxis()
# Create minor ticks at a increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
